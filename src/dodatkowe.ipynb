Za pomoca jednej z poznanych metod zaimplementuj generator liczb losowych z rozkÅ‚adu Poissona

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def generuj(lam, size):
    zmienne = np.zeros(size)
    for s in range(size):
        n = 1
        a = 1
        while True:
            U = np.random.uniform(0, 1)
            a *= U
            if a >= np.exp(-lam):
                n += 1
            else:
                zmienne[s] = n - 1
                break
    return zmienne
```

```python
lam, size = 1, 100
X = generuj(lam, size)
sns.kdeplot(X, label="RozkÅ‚ad wygenerowanych zmiennych", color="green")
sns.kdeplot(np.random.poisson(lam=lam, size=size), label="RozkÅ‚ad Poissona", color="red")
plt.title("PorÃ³wnanie rozkÅ‚adÃ³w")
plt.legend()
plt.show()
```
